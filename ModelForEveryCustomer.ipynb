```python
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt   
```

```python
#Read dataset
movies_with_customer_ratings_loaded = pd.read_csv('../dataset/prepared/movies_with_customer_ratings_and_all_features.csv')
```

```python
#So we dont have to reload the dataset every time
movies_with_customer_ratings = movies_with_customer_ratings_loaded
```

```python
movies_with_customer_ratings = movies_with_customer_ratings.drop(['Unnamed: 0'], axis=1)
movies_with_customer_ratings.head()
```

```python
#How many unique customers do we have?
number_unique_customers = movies_with_customer_ratings[['CustomerID']].groupby('CustomerID').count()
print(number_unique_customers)
```

```python
movies_with_customer_ratings = movies_with_customer_ratings.drop(['Budget', 'Genre', 'Original_language', 'Runtime'], axis=1)
```

```python
movies_with_customer_ratings.head()
```

```python
unique_users = movies_with_customer_ratings.CustomerID.unique()
#Calculate average rating for every customer
customer_averages = list()
counter = 0
for user in unique_users:
    movies_customer = movies_with_customer_ratings.loc[movies_with_customer_ratings['CustomerID'] == user]
    customer_mean = np.mean(movies_customer['Rating'])
    customer_averages.append(customer_mean)
    print(counter, ": ", customer_mean)
    counter += 1
sns.scatterplot(customer_averages)
```

```python
#Feature extraction
movies_30878 = movies_30878.drop(['Genre', 'Budget','Runtime','Original_language'], axis=1)
#Try to normalize month and day of rating as 31 and 1 are very close and not very far away?
movies_30878.head()
```

```python
data_corr = movies_30878.corr()
data_corr
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

target = movies_30878['Rating'].tolist()
data = movies_30878.drop(['Rating'], axis=1)

data_train_set_30878, data_test_set_30878, target_train_set_30878, target_test_set_30878 = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB

tree = DecisionTreeClassifier()

tree.fit(data_train_set_30878, target_train_set_30878)
target_predicted = tree.predict(data_test_set_30878)
accuracy_score(target_test_set_30878, target_predicted)
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(data_train_set_30878, target_train_set_30878)
target_predicted = svm.predict(data_test_set_30878)
accuracy_score(target_test_set_30878, target_predicted)
sns.scatterplot(data=target_predicted)
```

```python
#What is the average of all ratings for user 30878? Is it close to 4 (since all predicted ratings are 4)
average_rating_30878 = np.mean(movies_30878['Rating'])
print(average_rating_30878)
#Yes, close to 4. This makes it reasonable to believe that a simple average is as good as a prediction method as SVC and binary tree.
```
