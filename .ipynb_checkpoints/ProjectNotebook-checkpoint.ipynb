# Import packages

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

# Load dataset

```python
movies_dataframe = pd.read_csv('../dataset/movie_titles.txt', usecols=range(3), encoding='latin1', header=None) #Movie title may contain comma.
movies_dataframe.columns = ['MovieID','YearOfRelease','Title']
movies_dataframe.head()
#Scrape with genres?
```

```python
def getRatingsPathFromMovieID(id):
    basepath = '../dataset/training_set/mv_'
    fileEnding = '.txt'
    if(id<10):
        return basepath + '000000' + str(id) + fileEnding
    elif(id>=10 and id<100):
        return basepath + '00000' + str(id) + fileEnding
    elif(id>=100 and id<1000):
        return basepath + '0000' + str(id) + fileEnding
    elif(id>=1000 and id<10000):
        return basepath + '000' + str(id) + fileEnding
    elif(id>=10000 and id<100000):
        return basepath + '00' + str(id) + fileEnding
    else:
        return -1
```

```python
movie_rankings = list()
#for i in range(1,len(movies_dataframe.index)+1):
for i in range(1,25):
    path = getRatingsPathFromMovieID(i)
    try:
        movie_rankings.append(pd.read_csv(path, skiprows=[0], header=None))
        movie_rankings[i-1].columns = ['CustomerID', 'Rating', 'Date']
    except:
        print('Could not read txt file: ', path)
```

```python
df = movie_rankings[0]
df.head()
```

```python
print(len(movie_rankings))
```

```python
#Create a dataframe on the form CustomerID,MovieID,Rating,Date
user_movies = pd.DataFrame(columns=['CustomerID', 'MovieID', 'Rating', 'Date'])

for i in range(len(movie_rankings)):
    movieID = movies_dataframe.at[i,'MovieID']
    for j in range(len(movie_rankings[i])):
        customerID = movie_rankings[i].at[j,'CustomerID']
        rating = movie_rankings[i].at[j,'Rating']
        date = movie_rankings[i].at[j,'Date']
        #print('CustomerID: ', customerID, ' MovieID: ', movieID, ' Rating: ', rating, ' Date: ', date)
        user_movies = user_movies.append({'CustomerID':customerID, 'MovieID':movieID, 'Rating':rating, 'Date':date} , ignore_index=True)
user_movies
```

```python
user_movies_sorted = user_movies.sort_values(by=['CustomerID'])
user_movies_sorted
```

```python
customerID = 1333
single_customer_ratings = pd.DataFrame(columns=['CustomerID', 'MovieID', 'Rating', 'Date'])
for i in range(len(user_movies)):
    if(user_movies.at[i,'CustomerID'] == customerID):
        single_customer_ratings = single_customer_ratings.append(user_movies.loc[i])
single_customer_ratings
```

```python
single_customer_ratings_with_movies = pd.merge(user_movies, movies_dataframe[['MovieID', 'Title']], on='MovieID' )
single_customer_ratings_with_movies.Rating = single_user_ratings_with_movies.Rating.astype(int)
#single_user_ratings_with_movies.head()
customer_movie_ratings = pd.pivot_table(single_customer_ratings_with_movies, index='CustomerID', columns='Title', values='Rating')
customer_movie_ratings
```

```python
sns.heatmap(customer_movie_ratings)
```

```python
#Use only users who have rated M=5(?) movies
#Use only movies that have been rated N=20(?) times
#This is get a good basis on which to group users.

customer_movie_ratings['Count'] = customer_movie_ratings.count(axis=1)
customer_movie_ratings
```

```python
user_movies_sorted = user_movies.sort_values(by=['CustomerID'])
user_movies_sorted

customer_movie_ratings_sorted = customer_movie_ratings.sort_values(by=['Count'])
customer_movie_ratings_sorted
```
